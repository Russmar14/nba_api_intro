```python
from nba_api.stats.endpoints import playercareerstats 
#create obj 'jokic' to reference his stats, use method calling his player id 
jokic = playercareerstats.PlayerCareerStats(player_id = '203999')
#get it as a pandas dataframe
jokic.get_data_frames()[0]
#jokic, above, is a playerCareerStats (class) object. We called the method get_data_frames() to grab the first df
```

```python
#Let's get player and team ids
from nba_api.stats.static import teams

# get_teams returns a list of 30 dictionaries, each an NBA team.
nba_teams = teams.get_teams()
print("Number of teams fetched: {}".format(len(nba_teams)))
nba_teams[:3]

#okay now players
from nba_api.stats.static import players

# get_players returns a list of dictionaries, each representing a player.
nba_players = players.get_players()
print("Number of players fetched: {}".format(len(nba_players)))
nba_players[:5]
```

```python
#To search for player (or even team), use dictionary comprehensions:
big_fundamental = [player for player in nba_players if player["full_name"] == "Tim Duncan"][0]
big_fundamental['id']

```

```python
#Okay to put this together, let's call up Anthony Edwards' stats using his name and player id, from scratch
#import players, create list of dictionaries, each containing info on that player, including their player id, name, etc
from nba_api.stats.static import players
all_players = players.get_players()
#Okay now let's grab that dictionary for Joker using his name
#we grab index [0] because this returns a list holdin the dictionary as its only element
anthony_edwards_dict = [player for player in all_players if player['full_name'] == "Anthony Edwards"][0]
#Now we go back to calling playercareerstats and call the 'id' key from this dictionary
edwards = playercareerstats.PlayerCareerStats(player_id = anthony_edwards_dict['id'])
#get it as a pandas dataframe
edwards_df = edwards.get_data_frames()[0]
edwards_df
```

```python
#Okay now let's play around slightly to get used to manipulating stats with dataframes
#Let's pull a subset of rows: to pull only the full seasons, we'll try using season id
full_seasons = edwards_df[edwards_df['SEASON_ID'] < '2023-23']
#full_seasons
#How about a single column: to pull just the last few rows. Using .loc method
some_stats = edwards_df.loc[:,'AST':'PTS'] 
some_stats
```
